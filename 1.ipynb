```python
plt.bar([1.1,2.2,3.3,4.4],[10,20,30,40],label='BarPlot',width=0.3,color='r')
plt.xticks(rotation=45,color='b')
plt.yticks(rotation=45,color='b')
plt.xlabel('X',color='b')
plt.ylabel('Y',color='b',rotation=90)
plt.title('Y/X')
plt.legend()
plt.show()
```
